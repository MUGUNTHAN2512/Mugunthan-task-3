```python
for i in cont_cols:
    plt.figure(figsize=(10,5))
    plt.style.use("tableau-colorblind10")
    sns.distplot(sample_df[i], bins=30, color="r")
    plt.xlabel(i)
    plt.title("Frequency Distribution of"+i)
    plt.xticks(rotation=45)
    plt.show()
```

```python
for i in cat_cols:
    plt.figure(figsize=(8,5))
    plt.xlabel(i)
    plt.title("Count Distribution of"+i)
    plt.xticks(rotation=0)
    plt.grid(True)
    plt.show()
```

```python
for i in cont_cols:
    plt.figure(figsize=(10,5))
    sns.boxplot(sample_df[i])
    plt.xlabel(i)
    plt.title("Statistical Distribution of"+i)
    plt.xticks(rotation=0)
    plt.show()
```

```python
for x in cont_cols:
    for y in cont_cols:
        plt.figure(figsize=(10,5))
        if x!=y:
           plt.xlabel(x)
           plt.title("Relationship of "+x+" Vs +y")
           plt.xticks(rotation=0)
           plt.grid(True)
           plt.show()
```
